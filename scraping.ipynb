First let us collect Box office collection data from Bollywood Hungama website (because IMDb doesnt have it for may movies)

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import random
from fake_useragent import UserAgent
```

```python
df = pd.read_csv('movie_data.csv')
ua = UserAgent()
def get_bh_url(movie_name):
    url = f'https://www.bollywoodhungama.com/movie/{movie_name.replace(' ', '-').lower()}/box-office/'
    return url

def get_lifetime_collection(bh_url):
    headers = {'User-Agent':ua.random}
    try:
        response = requests.get(bh_url, headers=headers, timeout=10)
        soup = BeautifulSoup(response.text, 'html.parser')
        cell = soup.find_all('td', class_='lifetime')
        if cell:
            return cell[1].get_text(strip=True)
    except Exception as e:
        print(f"[ERROR] LifeTime collection from {bh_url}: {e}")
        return None

if 'bh_url' not in df.columns:
    df['bh_url'] = None
if 'box_office' not in df.columns:
    df['box_office'] = None

for i in range(101, 500):
    name = df.loc[i, 'Title']
    if pd.isna(df.loc[i, 'bh_url']):
        print(f"Searching bh url for {name}")
        url = get_bh_url(name)
        df.loc[i, 'bh_url'] = url
    bh_url = df.loc[i, 'bh_url']
    if pd.notna(bh_url) and pd.isna(df.loc[i, 'box_office']):
        print(f"Scraping box office collection from {bh_url}")
        amount = get_lifetime_collection(bh_url)
        df.loc[i, 'box_office'] = amount
        time.sleep(random.uniform(1.5, 3.5))
    

df.to_csv('movies_with_box_office.csv', index=False)
print('Complete')
```

Now scraping Director, Main Actors and writers from IMDb

```python
import numpy as np
def getDirector(data):
    try:
        return data['director'][0]['name']
    except Exception as e:
        print(f"[ERROR] {e} occured")
        return np.nan
def getActors(data):
    try:
        actor_cnt = len(data['actor'])
        actors = data['actor'][0]['name']
        for j in range(1, actor_cnt):
            actors += '+'
            actors += data['actor'][j]['name']
        return actors
    except Exception as e:
        print(f"[ERROR] {e} occured")
        return np.nan
def getWriters(data):
    try:
        writer_cnt = len(data['creator'])
        writers = data['creator'][1]['name']
        for j in range(2, writer_cnt):
            writers += '+'
            writers += data['creator'][j]['name']
        return writers
    except Exception as e:
        print(f"[ERROR] {e} occured")
        return np.nan

def getDuration(data):
    try:
        return data['duration']
    except Exception as e:
        print(f"[ERROR] {e} occured")
        return np.nan
def getRating(data):
    try:
        return float(data['aggregateRating']['ratingValue'])
    except:
        return np.nan
```

```python
if 'director' not in df.columns:
    df['director'] = None
if 'actors' not in df.columns:
    df['actors'] = None
if 'writers' not in df.columns:
    df['writers'] = None
if 'duration' not in df.columns:
    df['duration'] = None
if 'rating' not in df.columns:
    df['rating'] = None
```
